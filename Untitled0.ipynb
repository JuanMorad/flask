```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow # Importa esto para mostrar imágenes en Colab

# Define la función change_back
def change_back(background, img):
    """ Una función que reemplaza el fondo negro de las imágenes por otra imagen.
    Acepta dos argumentos: la imagen de fondo (background) y la propia imagen (img).
    La función comenzará desde la esquina superior izquierda de la imagen.
    Buscará todos los píxeles negros (con el valor 0) y los reemplazará por píxeles de la imagen de fondo."""

    x, y = 0, 0
    background = cv2.resize(background, (img.shape[1], img.shape[0]), interpolation = cv2.INTER_AREA)
    res = np.copy(background)
    place = res[y: y + img.shape[0], x: x + img.shape[1]]
    a = img[..., 3:].repeat(3, axis=2).astype('uint16')
    place[...] = (place.astype('uint16') * (255 - a) // 255) + img[..., :3].astype('uint16') * a // 255
    return res

# Carga la imagen a la que quieres cambiar el fondo (asegúrate de que tenga canal alfa)
# Reemplaza 'imagen_con_transparencia.png' con el nombre de tu archivo.
# Debes cargar esta imagen con IMREAD_UNCHANGED para leer el canal alfa.
imagen_a_procesar = cv2.imread('carro12.png', cv2.IMREAD_UNCHANGED)

# Carga la imagen que usarás como fondo
# Reemplaza 'fondo.jpg' con el nombre de tu archivo.
imagen_fondo = cv2.imread('leon.jpg')

# Verifica si las imágenes se cargaron correctamente
if imagen_a_procesar is None:
    print("Error al cargar la imagen a procesar.")
elif imagen_fondo is None:
    print("Error al cargar la imagen de fondo.")
else:
    # Llama a la función change_back
    imagen_con_nuevo_fondo = change_back(imagen_fondo, imagen_a_procesar)

    # Muestra la imagen resultante
    cv2_imshow(imagen_con_nuevo_fondo)
```
